### Problema de Negócio: Previsão de Demanda para Otimização de Estoque em E-commerce

#### Descrição Detalhada do Problema
Uma empresa de e-commerce de eletrônicos, "TechGlow", enfrenta desafios significativos na gestão de seu estoque. Atualmente, a TechGlow sofre com a superprodução de produtos de baixa demanda, resultando em custos de armazenamento elevados e obsolescência de estoque. Concomitantemente, há a falta de produtos populares, levando a perdas de vendas e insatisfação do cliente devido à indisponibilidade de itens desejados. A abordagem atual de previsão de demanda é baseada em métodos estatísticos simples e experiência de gerentes, que se mostram ineficazes diante da complexidade e sazonalidade do mercado de eletrônicos. Essa ineficiência resulta em uma perda estimada de 15% da receita anual devido a vendas perdidas e 10% em custos adicionais de estoque.

#### Relevância para o Cenário Empresarial
A otimização da previsão de demanda é crucial para a TechGlow. Uma previsão mais precisa permitirá à empresa:
*   **Reduzir Custos Operacionais:** Minimizar o excesso de estoque diminuirá os custos de armazenamento, manuseio e o risco de obsolescência.
*   **Aumentar a Receita:** Garantir a disponibilidade de produtos de alta demanda evitará perdas de vendas e aproveitará oportunidades de mercado.
*   **Melhorar a Satisfação do Cliente:** Clientes encontrarão os produtos que desejam prontamente, aumentando a lealdade e a reputação da marca.
*   **Otimizar o Capital de Giro:** Liberar capital que estaria preso em estoque parado, permitindo investimentos em outras áreas do negócio.

#### Objetivos SMART do Projeto
O projeto de Data Science e IA visa abordar este problema com os seguintes objetivos:

*   **Específico:** Desenvolver e implementar um modelo de previsão de demanda baseado em IA para produtos de eletrônicos na plataforma TechGlow.
*   **Mensurável:** Reduzir a taxa de *stockout* (falta de estoque) em 20% e diminuir o excesso de estoque (itens com mais de 180 dias em armazém) em 15% para os 100 produtos mais vendidos.
*   **Atingível:** Utilizar dados históricos de vendas, promoções, sazonalidade e tendências de mercado para treinar o modelo, com a infraestrutura de dados existente e uma equipe de ciência de dados dedicada.

Realizar uma análise exploratória completa, incluindo estatísticas descritivas, criação de visualizações relevantes para identificar padrões, tendências, anomalias e outliers nos dados, e extrair insights de negócio que guiarão as próximas etapas.


**Reasoning**:
The first step of the subtask is to create a synthetic DataFrame with specific columns and representative data. I will use pandas and numpy to generate this data.



```python
import pandas as pd
import numpy as np

# Set a random seed for reproducibility
np.random.seed(42)

# Define the number of days for the synthetic data
num_days = 365 * 2  # 2 years of data
start_date = pd.to_datetime('2022-01-01')

# Generate 'Data da Transação'
transaction_dates = pd.date_range(start=start_date, periods=num_days, freq='D')

# Generate 'ID do Produto/SKU' and 'Categoria do Produto'
product_categories = ['Eletrônicos', 'Vestuário', 'Alimentos', 'Livros', 'Casa e Decoração']
products_per_category = 10

product_skus = []
product_category_map = {}
for category in product_categories:
    for i in range(products_per_category):
        sku = f'{category[:3].upper()}-{i:03d}'
        product_skus.append(sku)
        product_category_map[sku] = category

# Simulate daily transactions (more transactions on some days)
num_transactions = num_days * 5 # Average 5 transactions per day

data = {
    'Data da Transação': np.random.choice(transaction_dates, size=num_transactions),
    'ID do Produto/SKU': np.random.choice(product_skus, size=num_transactions),
    'Quantidade Vendida': np.random.randint(1, 15, size=num_transactions), # Quantity between 1 and 14
    'Preço Unitário de Venda': np.round(np.random.uniform(10, 500, size=num_transactions), 2),
    'Promoção/Desconto': np.random.choice([0, 0.1, 0.25], size=num_transactions, p=[0.7, 0.2, 0.1]) # 70% no discount, 20% 10%, 10% 25%
}

df = pd.DataFrame(data)

# Add 'Categoria do Produto' based on 'ID do Produto/SKU'
df['Categoria do Produto'] = df['ID do Produto/SKU'].map(product_category_map)

# Ensure 'Data da Transação' is sorted for time series analysis
df = df.sort_values(by='Data da Transação').reset_index(drop=True)

print("Synthetic DataFrame created successfully.")
print(df.head())
```

**Reasoning**:
The next step is to convert the 'Data da Transação' column to datetime type (already done as per `df.info()`), set it as the DataFrame index, and then plot the total 'Quantidade Vendida' over time to identify trends and seasonality.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure 'Data da Transação' is datetime and set as index
df['Data da Transação'] = pd.to_datetime(df['Data da Transação'])
df_indexed = df.set_index('Data da Transação')

# Resample the data to daily frequency and sum 'Quantidade Vendida'
daily_sales = df_indexed['Quantidade Vendida'].resample('D').sum()

# Plot total 'Quantidade Vendida' over time
plt.figure(figsize=(16, 6))
plt.plot(daily_sales, color='skyblue')
plt.title('Total Quantidade Vendida ao Longo do Tempo (Diário)')
plt.xlabel('Data da Transação')
plt.ylabel('Quantidade Vendida Total')
plt.grid(True)
plt.tight_layout()
plt.show()

print("Plotting total 'Quantidade Vendida' over time completed.")
```

**Reasoning**:
I will now create histograms and box plots for the numerical variables 'Quantidade Vendida' and 'Preço Unitário de Venda' to visualize their distributions and identify potential outliers, as per the instructions.



```python
plt.figure(figsize=(15, 6))

# Histograma para 'Quantidade Vendida'
plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
sns.histplot(df['Quantidade Vendida'], kde=True, bins=range(1, df['Quantidade Vendida'].max() + 2), color='darkorange')
plt.title('Distribuição da Quantidade Vendida')
plt.xlabel('Quantidade Vendida')
plt.ylabel('Frequência')

# Box Plot para 'Quantidade Vendida'
plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
sns.boxplot(y=df['Quantidade Vendida'], color='lightcoral')
plt.title('Box Plot da Quantidade Vendida')
plt.ylabel('Quantidade Vendida')

plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 6))

# Histograma para 'Preço Unitário de Venda'
plt.subplot(1, 2, 1)
sns.histplot(df['Preço Unitário de Venda'], kde=True, bins=30, color='darkgreen')
plt.title('Distribuição do Preço Unitário de Venda')
plt.xlabel('Preço Unitário de Venda')
plt.ylabel('Frequência')

# Box Plot para 'Preço Unitário de Venda'
plt.subplot(1, 2, 2)
sns.boxplot(y=df['Preço Unitário de Venda'], color='lightseagreen')
plt.title('Box Plot do Preço Unitário de Venda')
plt.ylabel('Preço Unitário de Venda')

plt.tight_layout()
plt.show()

print("Histograms and Box Plots for 'Quantidade Vendida' and 'Preço Unitário de Venda' created.")
```

**Reasoning**:
I will now analyze the distribution of 'Quantidade Vendida' by 'Categoria do Produto' to identify the most sold categories and any significant differences in demand between them. I will use a bar plot to show the total quantity sold per category and a box plot to visualize the distribution of sales within each category.



```python
plt.figure(figsize=(15, 6))

# Bar plot of total 'Quantidade Vendida' per 'Categoria do Produto'
plt.subplot(1, 2, 1)
category_sales = df.groupby('Categoria do Produto')['Quantidade Vendida'].sum().sort_values(ascending=False)
sns.barplot(x=category_sales.index, y=category_sales.values, palette='viridis')
plt.title('Total Quantidade Vendida por Categoria do Produto')
plt.xlabel('Categoria do Produto')
plt.ylabel('Quantidade Vendida Total')
plt.xticks(rotation=45, ha='right')

# Box plot of 'Quantidade Vendida' distribution per 'Categoria do Produto'
plt.subplot(1, 2, 2)
sns.boxplot(x='Categoria do Produto', y='Quantidade Vendida', data=df, palette='viridis')
plt.title('Distribuição da Quantidade Vendida por Categoria do Produto')
plt.xlabel('Categoria do Produto')
plt.ylabel('Quantidade Vendida')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()

print("Plots for 'Quantidade Vendida' by 'Categoria do Produto' created.")
```

**Reasoning**:
The previous code generated a `FutureWarning` because `palette` was used without assigning a `hue` argument in `sns.barplot` and `sns.boxplot`. I will fix this by explicitly setting `hue` to the same column as `x` and adding `legend=False` to align with the recommended usage in newer Seaborn versions.



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 6))

# Bar plot of total 'Quantidade Vendida' per 'Categoria do Produto'
plt.subplot(1, 2, 1)
category_sales = df.groupby('Categoria do Produto')['Quantidade Vendida'].sum().sort_values(ascending=False)
sns.barplot(x=category_sales.index, y=category_sales.values, palette='viridis', hue=category_sales.index, legend=False)
plt.title('Total Quantidade Vendida por Categoria do Produto')
plt.xlabel('Categoria do Produto')
plt.ylabel('Quantidade Vendida Total')
plt.xticks(rotation=45, ha='right')

# Box plot of 'Quantidade Vendida' distribution per 'Categoria do Produto'
plt.subplot(1, 2, 2)
sns.boxplot(x='Categoria do Produto', y='Quantidade Vendida', data=df, palette='viridis', hue='Categoria do Produto', legend=False)
plt.title('Distribuição da Quantidade Vendida por Categoria do Produto')
plt.xlabel('Categoria do Produto')
plt.ylabel('Quantidade Vendida')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()

print("Plots for 'Quantidade Vendida' by 'Categoria do Produto' created.")
```

I will now investigate the impact of 'Promoção/Desconto' on 'Quantidade Vendida' using a box plot, as requested by the instructions. This will allow us to compare the distribution of sales quantities under different discount levels.



```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Promoção/Desconto', y='Quantidade Vendida', data=df, palette='viridis')
plt.title('Impacto da Promoção/Desconto na Quantidade Vendida')
plt.xlabel('Nível de Promoção/Desconto')
plt.ylabel('Quantidade Vendida')
plt.xticks(ticks=[0, 1, 2], labels=['Sem Desconto (0%)', '10% de Desconto', '25% de Desconto'])
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()

print("Box plot for 'Promoção/Desconto' vs 'Quantidade Vendida' created.")
```

**Reasoning**:
The next step is to create 'lag features' for 'Quantidade Vendida'. I will sort the DataFrame by 'Data da Transação' and 'ID do Produto/SKU', then calculate the lag for 1 day, 7 days, and 30 days for each product's quantity sold. This will capture past sales patterns, which are crucial for demand forecasting.



```python
df_sorted = df.sort_values(by=['Data da Transação', 'ID do Produto/SKU']).copy()

# Create lag features for 'Quantidade Vendida' grouped by 'ID do Produto/SKU'
df_sorted['Lag_1_Dia_Qtde'] = df_sorted.groupby('ID do Produto/SKU')['Quantidade Vendida'].shift(1)
df_sorted['Lag_7_Dias_Qtde'] = df_sorted.groupby('ID do Produto/SKU')['Quantidade Vendida'].shift(7)
df_sorted['Lag_30_Dias_Qtde'] = df_sorted.groupby('ID do Produto/SKU')['Quantidade Vendida'].shift(30)

df = df_sorted # Update the main dataframe

print("Lag features created successfully.")
print(df[['Data da Transação', 'ID do Produto/SKU', 'Quantidade Vendida', 'Lag_1_Dia_Qtde', 'Lag_7_Dias_Qtde', 'Lag_30_Dias_Qtde']].head(10))
```

**Reasoning**:
I will now create 'rolling window features' for 'Quantidade Vendida'. Specifically, I will calculate the rolling mean of sales for the last 3 and 7 days for each product SKU. This captures recent sales trends and smooths out short-term fluctuations, which is valuable for demand forecasting.



```python
df_sorted = df.sort_values(by=['Data da Transação', 'ID do Produto/SKU']).copy()

# Create rolling window features for 'Quantidade Vendida' grouped by 'ID do Produto/SKU'
df_sorted['Rolling_Mean_3_Dias_Qtde'] = df_sorted.groupby('ID do Produto/SKU')['Quantidade Vendida'].transform(lambda x: x.rolling(window=3, min_periods=1).mean())
df_sorted['Rolling_Mean_7_Dias_Qtde'] = df_sorted.groupby('ID do Produto/SKU')['Quantidade Vendida'].transform(lambda x: x.rolling(window=7, min_periods=1).mean())

df = df_sorted # Update the main dataframe

print("Rolling window features created successfully.")
print(df[['Data da Transação', 'ID do Produto/SKU', 'Quantidade Vendida', 'Rolling_Mean_3_Dias_Qtde', 'Rolling_Mean_7_Dias_Qtde']].head(10))
```

I will fill these NaNs with 0, assuming that the absence of prior data implies no sales for those specific periods, which is a common approach for time-series features.



```python
print("Missing values before handling:\n", df.isnull().sum()[df.isnull().sum() > 0])

# Fill NaN values generated by lag and rolling features with 0
# This assumes that missing past data implies zero sales for that period
lag_rolling_features = [
    'Lag_1_Dia_Qtde',
    'Lag_7_Dias_Qtde',
    'Lag_30_Dias_Qtde',
    'Rolling_Mean_3_Dias_Qtde',
    'Rolling_Mean_7_Dias_Qtde'
]

for col in lag_rolling_features:
    if col in df.columns:
        df[col] = df[col].fillna(0)

print("\nMissing values after handling:\n", df.isnull().sum()[df.isnull().sum() > 0])
print("Missing values in lag and rolling features have been filled with 0.")
```

### Justificativa das Escolhas de Engenharia e Pré-processamento de Features

Nesta etapa, realizamos uma série de transformações e criações de novas variáveis (feature engineering) e preparamos os dados (pré-processamento) para otimizar o desempenho de modelos de previsão de demanda. Abaixo, detalhamos e justificamos cada escolha:

1.  **Extração de Recursos Baseados em Tempo (`Ano`, `Mes`, `Dia do Mes`, `Dia da Semana`, `Dia do Ano`, `Semana do Ano`, `Fim de Semana`):**
    *   **Justificativa:** A demanda por produtos é intrinsecamente ligada ao tempo. Padrões sazonais (mensais, anuais), variações diárias (dias da semana vs. fins de semana) e eventos específicos ao longo do ano são fatores cruciais. Ao extrair essas features, permitimos que o modelo capture tendências temporais, sazonalidades e ciclos que influenciam as vendas. Por exemplo, saber se é um fim de semana pode indicar um pico de compras online ou, inversamente, uma queda em certos segmentos de negócio.

2.  **Criação de `Valor Total da Venda`:**
    *   **Justificativa:** Embora `Quantidade Vendida` seja nossa variável alvo principal, `Valor Total da Venda` (`Quantidade Vendida` * `Preço Unitário de Venda`) representa a receita gerada por transação. Pode ser útil para análises de rentabilidade e para modelos que buscam otimizar o valor total das vendas, não apenas a quantidade. Além disso, pode ser utilizada como uma feature auxiliar para modelos de demanda mais complexos, ou mesmo como target secundário.

3.  **Criação de `Lag Features` (`Lag_1_Dia_Qtde`, `Lag_7_Dias_Qtde`, `Lag_30_Dias_Qtde`):**
    *   **Justificativa:** As vendas passadas são um dos preditores mais fortes das vendas futuras. As `lag features` capturam a dependência temporal da série, permitindo que o modelo "lembre" o histórico de vendas de um produto específico. Um lag de 1 dia reflete a demanda imediatamente anterior, enquanto lags de 7 e 30 dias ajudam a identificar padrões semanais e mensais, respectivamente, essenciais para prever a demanda futura com base no comportamento recente e recorrente.

4.  **Criação de `Rolling Window Features` (`Rolling_Mean_3_Dias_Qtde`, `Rolling_Mean_7_Dias_Qtde`):**
    *   **Justificativa:** As médias móveis suavizam as flutuações de curto prazo e revelam tendências de vendas mais estáveis ao longo do tempo. Uma média móvel de 3 dias captura a tendência de vendas muito recente, enquanto uma de 7 dias pode capturar padrões semanais, atenuando o ruído diário. Essas features são valiosas para o modelo entender o ritmo médio de vendas de um produto, sendo menos sensíveis a picos ou quedas isoladas que as `lag features` puras.

5.  **One-Hot Encoding da `Categoria do Produto`:**
    *   **Justificativa:** `Categoria do Produto` é uma variável categórica nominal. Modelos de machine learning geralmente requerem entradas numéricas. O one-hot encoding converte cada categoria em uma nova coluna binária (0 ou 1), sem implicar qualquer ordem ou hierarquia entre as categorias. Isso permite que o modelo trate cada categoria como uma entidade distinta e aprenda a influência de cada uma na demanda, evitando que o modelo interprete uma ordem artificial se fosse usada codificação ordinal.

6.  **Tratamento de Valores Ausentes em `Lag Features` e `Rolling Window Features`:**
    *   **Justificativa:** `Lag features` e `rolling window features` naturalmente geram valores `NaN` no início da série temporal, pois não há dados anteriores para calcular o lag ou a média. Preencher esses `NaNs` com 0 é uma estratégia comum e justificável para a previsão de demanda. Isso implica que, para os primeiros pontos da série de um produto, não havia vendas passadas (ou dados suficientes) para informar esses recursos, o que é uma informação válida para o modelo.

7.  **Normalização de Colunas Numéricas com `StandardScaler`:**
    *   **Justificativa:** `StandardScaler` transforma os dados para que tenham média 0 e desvio padrão 1 (distribuição normal padrão). Isso é crucial para muitos algoritmos de machine learning (especialmente aqueles baseados em distância, como SVMs ou redes neurais, e métodos baseados em gradiente) pois evita que features com escalas maiores dominem o processo de otimização. Ao normalizar `Preço Unitário de Venda`, `Valor Total da Venda`, e todas as `lag` e `rolling window features`, garantimos que todas as variáveis numéricas contribuam igualmente para o treinamento do modelo, melhorando a convergência e a performance.

**Reasoning**:

```python
print("Making predictions with Linear Regression...")
lin_reg_preds = lin_reg.predict(X_test)
print("Predictions made with Linear Regression.")

print("\nMaking predictions with RandomForest Regressor...")
rf_reg_preds = rf_reg.predict(X_test)
print("Predictions made with RandomForest Regressor.")

print("\nMaking predictions with XGBoost Regressor...")
xgb_reg_preds = xgb_reg.predict(X_test)
print("Predictions made with XGBoost Regressor.")

print("All models have made predictions successfully.")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Evaluate Linear Regression
mae_lin_reg = mean_absolute_error(y_test, lin_reg_preds)
mse_lin_reg = mean_squared_error(y_test, lin_reg_preds)
rmse_lin_reg = np.sqrt(mse_lin_reg)

print("Linear Regression Metrics:")
print(f"  MAE: {mae_lin_reg:.4f}")
print(f"  MSE: {mse_lin_reg:.4f}")
print(f"  RMSE: {rmse_lin_reg:.4f}")

# Evaluate RandomForest Regressor
mae_rf_reg = mean_absolute_error(y_test, rf_reg_preds)
mse_rf_reg = mean_squared_error(y_test, rf_reg_preds)
rmse_rf_reg = np.sqrt(mse_rf_reg)

print("\nRandomForest Regressor Metrics:")
print(f"  MAE: {mae_rf_reg:.4f}")
print(f"  MSE: {mse_rf_reg:.4f}")
print(f"  RMSE: {rmse_rf_reg:.4f}")

# Evaluate XGBoost Regressor
mae_xgb_reg = mean_absolute_error(y_test, xgb_reg_preds)
mse_xgb_reg = mean_squared_error(y_test, xgb_reg_preds)
rmse_xgb_reg = np.sqrt(mse_xgb_reg)

print("\nXGBoost Regressor Metrics:")
print(f"  MAE: {mae_xgb_reg:.4f}")
print(f"  MSE: {mse_xgb_reg:.4f}")
print(f"  RMSE: {rmse_xgb_reg:.4f}")

print("All models evaluated successfully.")
```

```python
import pandas as pd
import numpy as np
from datetime import timedelta
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

# --- REGENERATING DF FROM SCRATCH TO ENSURE CORRECT SCALING ---
# Recreate synthetic data (from cell e8cde0ab)
np.random.seed(42)
num_days = 365 * 2
start_date = pd.to_datetime('2022-01-01')
transaction_dates = pd.date_range(start=start_date, periods=num_days, freq='D')
product_categories = ['Eletrônicos', 'Vestuário', 'Alimentos', 'Livros', 'Casa e Decoração']
products_per_category = 10
product_skus = []
product_category_map = {}
for category in product_categories:
    for i in range(products_per_category):
        sku = f'{category[:3].upper()}-{i:03d}'
        product_skus.append(sku)
        product_category_map[sku] = category
num_transactions = num_days * 5
data = {
    'Data da Transação': np.random.choice(transaction_dates, size=num_transactions),
    'ID do Produto/SKU': np.random.choice(product_skus, size=num_transactions),
    'Quantidade Vendida': np.random.randint(1, 15, size=num_transactions),
    'Preço Unitário de Venda': np.round(np.random.uniform(10, 500, size=num_transactions), 2),
    'Promoção/Desconto': np.random.choice([0, 0.1, 0.25], size=num_transactions, p=[0.7, 0.2, 0.1])
}
df_raw = pd.DataFrame(data)
df_raw['Categoria do Produto'] = df_raw['ID do Produto/SKU'].map(product_category_map)
df_raw = df_raw.sort_values(by='Data da Transação').reset_index(drop=True)

# Feature Engineering (from previous cells)
df_raw['Ano'] = df_raw['Data da Transação'].dt.year
df_raw['Mes'] = df_raw['Data da Transação'].dt.month
df_raw['Dia do Mes'] = df_raw['Data da Transação'].dt.day
df_raw['Dia da Semana'] = df_raw['Data da Transação'].dt.dayofweek
df_raw['Dia do Ano'] = df_raw['Data da Transação'].dt.dayofyear
df_raw['Semana do Ano'] = df_raw['Data da Transação'].dt.isocalendar().week.astype(int)
df_raw['Fim de Semana'] = df_raw['Data da Transação'].dt.dayofweek >= 5
df_raw['Valor Total da Venda'] = df_raw['Quantidade Vendida'] * df_raw['Preço Unitário de Venda']

df_raw_sorted = df_raw.sort_values(by=['Data da Transação', 'ID do Produto/SKU']).copy()
df_raw_sorted['Lag_1_Dia_Qtde'] = df_raw_sorted.groupby('ID do Produto/SKU')['Quantidade Vendida'].shift(1)
df_raw_sorted['Lag_7_Dias_Qtde'] = df_raw_sorted.groupby('ID do Produto/SKU')['Quantidade Vendida'].shift(7)
df_raw_sorted['Lag_30_Dias_Qtde'] = df_raw_sorted.groupby('ID do Produto/SKU')['Quantidade Vendida'].shift(30)
df_raw_sorted['Rolling_Mean_3_Dias_Qtde'] = df_raw_sorted.groupby('ID do Produto/SKU')['Quantidade Vendida'].transform(lambda x: x.rolling(window=3, min_periods=1).mean())
df_raw_sorted['Rolling_Mean_7_Dias_Qtde'] = df_raw_sorted.groupby('ID do Produto/SKU')['Quantidade Vendida'].transform(lambda x: x.rolling(window=7, min_periods=1).mean())
df_raw = df_raw_sorted

# One-hot encoding
df_raw = pd.get_dummies(df_raw, columns=['Categoria do Produto'], prefix='Categoria', dtype=int)

# Handle NaNs for lag/rolling features
lag_rolling_features = [
    'Lag_1_Dia_Qtde',
    'Lag_7_Dias_Qtde',
    'Lag_30_Dias_Qtde',
    'Rolling_Mean_3_Dias_Qtde',
    'Rolling_Mean_7_Dias_Qtde'
]
for col in lag_rolling_features:
    if col in df_raw.columns:
        df_raw[col] = df_raw[col].fillna(0)

# --- CORRECT TRAIN/TEST SPLIT AND SCALING ---

# Define target variable (y) and features (X)
y = df_raw['Quantidade Vendida']

# Define features, dropping identifiers and the target, and 'Valor Total da Venda' as it's not a feature for prediction
X_full = df_raw.drop(columns=['Quantidade Vendida', 'Data da Transação', 'ID do Produto/SKU', 'Valor Total da Venda'])
X_full['Fim de Semana'] = X_full['Fim de Semana'].astype(int) # Ensure boolean is int

# Identify the split point for 80% training and 20% testing
split_point = int(len(df_raw) * 0.8)

X_train_raw = X_full.iloc[:split_point].copy() # Make copies to avoid SettingWithCopyWarning
X_test_raw = X_full.iloc[split_point:].copy()
y_train = y.iloc[:split_point]
y_test = y.iloc[split_point:]

# Identify numerical columns for scaling *from the unscaled data*
numerical_cols_for_scaling = [
    'Preço Unitário de Venda',
    'Promoção/Desconto',
    'Lag_1_Dia_Qtde',
    'Lag_7_Dias_Qtde',
    'Lag_30_Dias_Qtde',
    'Rolling_Mean_3_Dias_Qtde',
    'Rolling_Mean_7_Dias_Qtde'
]
numerical_cols_for_scaling = [col for col in numerical_cols_for_scaling if col in X_full.columns]

# Initialize and fit scaler on X_train_raw's numerical columns (UNSCALED data)
scaler = StandardScaler()
scaler.fit(X_train_raw[numerical_cols_for_scaling])

# Transform X_train_raw and X_test_raw
X_train = X_train_raw.copy()
X_test = X_test_raw.copy()
X_train[numerical_cols_for_scaling] = scaler.transform(X_train_raw[numerical_cols_for_scaling])
X_test[numerical_cols_for_scaling] = scaler.transform(X_test_raw[numerical_cols_for_scaling])

print("Data regenerated, features engineered, and scaled correctly for X_train and X_test.")
print(f"X_train shape: {X_train.shape}, X_test shape: {X_test.shape}")

# --- RETRAIN MODELS ---
rf_reg = RandomForestRegressor(random_state=42)
rf_reg.fit(X_train, y_train)
print("RandomForest Regressor model retrained on correctly scaled data.")

# --- GENERATE df_future (UNSCALED for price/promo init) ---
last_historical_date = df_raw['Data da Transação'].max()
future_days = 30
future_dates = pd.date_range(start=last_historical_date + timedelta(days=1), periods=future_days, freq='D')

# Get unique product SKUs and their one-hot encoded categories from the original (unscaled) df_raw
unique_products_info = df_raw[['ID do Produto/SKU', 'Categoria_Alimentos', 'Categoria_Casa e Decoração',
                       'Categoria_Eletrônicos', 'Categoria_Livros', 'Categoria_Vestuário']].drop_duplicates()

df_future_base = pd.DataFrame({
    'Data da Transação': np.repeat(future_dates, len(unique_products_info)),
    'ID do Produto/SKU': np.tile(unique_products_info['ID do Produto/SKU'], future_days)
})

df_future_base = pd.merge(df_future_base, unique_products_info, on='ID do Produto/SKU', how='left')

# --- FEATURE ENGINEERING FOR df_future_base ---
df_future_base['Ano'] = df_future_base['Data da Transação'].dt.year
df_future_base['Mes'] = df_future_base['Data da Transação'].dt.month
df_future_base['Dia do Mes'] = df_future_base['Data da Transação'].dt.day
df_future_base['Dia da Semana'] = df_future_base['Data da Transação'].dt.dayofweek
df_future_base['Dia do Ano'] = df_future_base['Data da Transação'].dt.dayofyear
df_future_base['Semana do Ano'] = df_future_base['Data da Transação'].dt.isocalendar().week.astype(int)
df_future_base['Fim de Semana'] = df_future_base['Data da Transação'].dt.dayofweek >= 5
df_future_base['Fim de Semana'] = df_future_base['Fim de Semana'].astype(int)

# --- FILL 'Preço Unitário de Venda' and 'Promoção/Desconto' in df_future with UNscaled historical averages ---
# Calculate unscaled historical means from df_raw
historical_product_unscaled_data = df_raw.groupby('ID do Produto/SKU')[['Preço Unitário de Venda', 'Promoção/Desconto']].mean().reset_index()

# Merge historical unscaled data. Since df_future_base does not yet have 'Preço Unitário de Venda' and 'Promoção/Desconto'
# these columns will be added directly from historical_product_unscaled_data without suffixes.
df_future_base = pd.merge(df_future_base, historical_product_unscaled_data, on='ID do Produto/SKU', how='left')

# --- Initialize lag and rolling window features in df_future_base to 0 ---
for col in lag_rolling_features:
    df_future_base[col] = 0.0

# --- ALIGN AND SCALE df_future_base for prediction ---
# Get columns from X_train (which are already scaled numerical features + one-hot categories + time features)
X_train_cols = X_train.columns.tolist()

# Create a temporary DataFrame for aligning and scaling
df_future_to_align_scale = df_future_base.copy()

# Ensure all X_train columns are present in df_future_to_align_scale, fill with 0 if missing
for col in X_train_cols:
    if col not in df_future_to_align_scale.columns:
        df_future_to_align_scale[col] = 0

# Drop extra columns not present in X_train
extra_cols = [col for col in df_future_to_align_scale.columns if col not in X_train_cols and col not in ['Data da Transação', 'ID do Produto/SKU']]
df_future_to_align_scale = df_future_to_align_scale.drop(columns=extra_cols)

# Ensure the order is exactly the same as X_train for prediction
df_future_aligned = df_future_to_align_scale[X_train_cols]

# Apply the *fitted* scaler to the selected numerical columns in df_future_aligned
df_future_aligned[numerical_cols_for_scaling] = scaler.transform(df_future_aligned[numerical_cols_for_scaling])

print("df_future_aligned prepared and scaled correctly for predictions.")
print(f"Shape of df_future_aligned: {df_future_aligned.shape}")
print("First 5 rows of df_future_aligned (scaled numerical columns):")
print(df_future_aligned[numerical_cols_for_scaling].head())

# --- MAKE PREDICTIONS ---
# Predict using the RandomForest Regressor on the scaled future data
df_future_base['Quantidade_Vendida_Prevista'] = rf_reg.predict(df_future_aligned)

# Ensure predictions are non-negative and round to integer quantity
df_future_base['Quantidade_Vendida_Prevista'] = df_future_base['Quantidade_Vendida_Prevista'].apply(lambda x: max(0, round(x, 0)))

# Make a copy for final predictions to be used in next steps
df_future_final_predictions = df_future_base.copy()

print("\nFuture demand predictions generated successfully on the re-processed data.")
print(df_future_final_predictions[['Data da Transação', 'ID do Produto/SKU', 'Preço Unitário de Venda', 'Quantidade_Vendida_Prevista']].head())

# --- Calculate Receita_Prevista_SKU using UNscaled prices and predicted quantities ---
df_future_final_predictions['Receita_Prevista_SKU'] = df_future_final_predictions['Quantidade_Vendida_Prevista'] * df_future_final_predictions['Preço Unitário de Venda']

# Aggregate to get 'Receita Prevista Total' daily
total_daily_predicted_revenue = df_future_final_predictions.groupby('Data da Transação')['Receita_Prevista_SKU'].sum().reset_index()

print("\nTotal daily predicted revenue calculated successfully with unscaled prices.")
print(total_daily_predicted_revenue.head())
```

Detalhar a arquitetura do projeto, incluindo o pipeline de dados. Discutir as estratégias para um possível deploy em nuvem, integração via APIs ou automações, e abordar a escalabilidade da solução para atender futuras demandas.


### 1. Arquitetura de Alto Nível da Solução de Previsão de Demanda

A arquitetura proposta para a solução de previsão de demanda da TechGlow é modular e escalável, projetada para cobrir todo o ciclo de vida dos dados e do modelo, desde a ingestão até a disponibilização das previsões para o negócio. Os principais componentes são:

*   **Fontes de Dados:** Representam os sistemas de origem que contêm as informações necessárias para a previsão de demanda. Incluem:
    *   **Sistemas de ERP/E-commerce:** Para dados transacionais (vendas, estoque, produtos, preços, promoções).
    *   **Sistemas de CRM:** Para dados de clientes e interações.
    *   **Fontes Externas:** APIs de dados climáticos, indicadores econômicos, calendários de feriados, dados de mídias sociais e de concorrência.

*   **Ingestão de Dados:** Responsável por coletar e transferir dados das fontes para o ambiente de processamento e armazenamento. Pode ser realizada via:
    *   **Batch Processing:** Para dados históricos e atualizações periódicas (e.g., diárias, semanais).
    *   **Real-time Streaming:** Para dados de eventos que exigem processamento imediato (e.g., novas vendas, mudanças de preço).

*   **Armazenamento de Dados (Data Lake/Data Warehouse):** Um repositório centralizado e escalável para todos os dados brutos e processados.
    *   **Data Lake (e.g., Amazon S3, Azure Data Lake Storage):** Armazena dados brutos e semi-estruturados em seu formato original, permitindo flexibilidade para futuras análises.
    *   **Data Warehouse (e.g., Amazon Redshift, Google BigQuery, Snowflake):** Armazena dados estruturados e limpos, otimizados para consultas analíticas e relatórios.

*   **Processamento de Dados (ETL/ELT):** Componentes que realizam a extração, transformação e carregamento dos dados.
    *   **Ferramentas de ETL/ELT:** Apache Spark, AWS Glue, Azure Data Factory, DBT (Data Build Tool) para limpeza, enriquecimento, agregação e engenharia de features.

*   **Treinamento do Modelo (Machine Learning Platform):** Ambiente para desenvolvimento, treinamento, validação e gerenciamento de modelos de ML.
    *   **Plataformas:** Amazon SageMaker, Google AI Platform, Azure Machine Learning. Permitem experimentação, versionamento de modelos e automação do pipeline de ML (MLOps).
    *   **Modelos:** RandomForest Regressor (escolhido), XGBoost, Prophet, etc.

*   **Disponibilização do Modelo (Model Deployment):** Mecanismos para colocar o modelo treinado em produção e gerar previsões.
    *   **Batch Prediction:** Para previsões periódicas (e.g., diárias, semanais) de todos os produtos.
    *   **Real-time API Endpoint:** Para previsões sob demanda para produtos específicos (e.g., ao consultar estoque de um item no e-commerce).

*   **Dashboard e Visualização:** Interface para apresentar as previsões e KPIs de forma clara e interativa para usuários de negócio.
    *   **Ferramentas:** Tableau, Power BI, Looker, Streamlit. Integram-se com o Data Warehouse ou com o serviço de disponibilização do modelo.

### 2. Detalhamento do Pipeline de Dados

O pipeline de dados é o coração da solução, garantindo que os dados fluam de forma eficiente, sejam transformados adequadamente e estejam disponíveis para treinamento e inferência do modelo. O fluxo seria o seguinte:

1.  **Extração (Extract):**
    *   **Fontes Internas:** Dados de vendas, estoque e produtos seriam extraídos diretamente dos sistemas ERP e da plataforma de e-commerce da TechGlow. Isso pode ser feito via conectores diretos a bancos de dados (e.g., PostgreSQL, MySQL, SQL Server) ou APIs internas.
    *   **Fontes Externas:** Dados de feriados, clima e indicadores econômicos seriam coletados de APIs públicas ou serviços de terceiros. Dados de redes sociais e concorrência poderiam ser obtidos via web scraping ou APIs específicas.
    *   **Ferramentas:** AWS Glue, Azure Data Factory, Apache Nifi, Airbyte, ou scripts Python agendados via Apache Airflow.

2.  **Carregamento Inicial (Load to Data Lake):**
    *   Os dados brutos extraídos, independentemente do formato (CSV, JSON, XML, Parquet), seriam carregados no Data Lake (e.g., Amazon S3, Azure Data Lake Storage) em sua forma original. Este passo garante que nenhum dado seja perdido e permite a retroanálise, se necessário.
    *   **Ferramentas:** Conectores de serviços de ingestão, AWS S3, Azure Blob Storage.

3.  **Transformação e Enriquecimento (Transform in Data Warehouse):**
    *   **Limpeza de Dados:** Tratamento de valores ausentes, correção de formatos, remoção de duplicatas.
    *   **Engenharia de Features:** Criação de variáveis temporais (ano, mês, dia da semana), lag features (vendas dos N dias anteriores), rolling window features (médias móveis de vendas), indicadores de promoção, etc.
    *   **Agregação:** Consolidação de dados transacionais para o nível de granularidade desejado para a previsão (e.g., vendas diárias por SKU e categoria).
    *   **Enriquecimento:** Junção de dados de diferentes fontes (e.g., combinar dados de vendas com informações de produtos e eventos externos).
    *   **Armazenamento de Dados Transformados:** Os dados limpos e transformados seriam carregados em um Data Warehouse (e.g., Google BigQuery, Amazon Redshift) para armazenamento estruturado e otimizado para consultas.
    *   **Ferramentas:** Apache Spark (via Databricks, AWS EMR), dbt (Data Build Tool), scripts Python/SQL orquestrados por Apache Airflow.

4.  **Treinamento do Modelo:**
    *   O Data Warehouse alimentaria diretamente o ambiente de treinamento do modelo.
    *   Os dados seriam divididos em conjuntos de treinamento e validação.
    *   O modelo (RandomForest Regressor, por exemplo) seria treinado e otimizado.
    *   **Ferramentas:** Amazon SageMaker, Google AI Platform, Azure Machine Learning, com scripts Python e bibliotecas como Scikit-learn, XGBoost.

5.  **Inferência e Disponibilização de Previsões:**
    *   **Geração de Features para Previsão:** Novas features (temporais, lag, rolling) seriam geradas para o período futuro a ser previsto, com base nos dados mais recentes disponíveis no Data Warehouse.
    *   **Predição:** O modelo treinado seria utilizado para gerar as previsões de demanda.
    *   **Armazenamento das Previsões:** As previsões geradas seriam armazenadas de volta no Data Warehouse ou em um banco de dados de alta performance (e.g., DynamoDB, Redis) otimizado para acesso rápido por APIs e dashboards.
    *   **Ferramentas:** Funções serverless (AWS Lambda, Azure Functions), Kubernetes (para endpoints de API), Apache Airflow (para orquestração de batch prediction).

Este pipeline garante a integridade, qualidade e relevância dos dados em todas as etapas, desde a coleta bruta até a disponibilização das previsões para consumo.

### 3. Estratégias para o Deploy do Modelo de Machine Learning

O deploy do modelo de Machine Learning é a etapa crucial que transforma a solução analítica em um componente operacional que entrega valor ao negócio. A escolha da estratégia de deploy dependerá dos requisitos de latência, volume de previsões e criticidade para o negócio da TechGlow.

#### A. Predições em Lote (Batch Prediction):

Esta abordagem é ideal para gerar previsões de demanda de forma periódica para um grande volume de produtos. É adequada para o planejamento de estoque que não exige atualizações em tempo real, como previsões diárias, semanais ou mensais para todos os SKUs.

*   **Fluxo de Operação:**
    1.  **Geração de Features:** Em um cronograma definido (e.g., uma vez por dia à noite), o pipeline de dados seria executado para coletar os dados mais recentes do Data Warehouse e gerar as features de entrada necessárias para o modelo (variáveis temporais, lags, rolling means, etc.) para o período futuro a ser previsto.
    2.  **Execução do Modelo:** O modelo treinado (RandomForest Regressor, neste caso) seria carregado e executaria previsões sobre o conjunto de dados preparado.
    3.  **Armazenamento de Previsões:** As previsões geradas seriam armazenadas em uma tabela específica no Data Warehouse (ou em um banco de dados otimizado para consultas) para consumo por outros sistemas ou pelo dashboard.
    4.  **Orquestração:** Ferramentas como Apache Airflow, AWS Step Functions ou Azure Data Factory seriam usadas para orquestrar e monitorar todo o processo, garantindo que as etapas sejam executadas em sequência e que falhas sejam tratadas.

*   **Vantagens:** Eficiente para grandes volumes de dados, menor custo operacional em comparação com tempo real, mais fácil de gerenciar e depurar.
*   **Desvantagens:** Latência nas previsões (previsões só são atualizadas no próximo ciclo do lote), não adequada para decisões que exigem informações instantâneas.

#### B. Predições em Tempo Real (Real-time Prediction via API):

Esta abordagem é indicada para cenários onde previsões instantâneas são necessárias, como ao consultar a disponibilidade de estoque de um produto no site em tempo de compra, ou para ajustar dinamicamente preços ou promoções. Um endpoint de API seria exposto para que outros sistemas possam solicitar previsões para produtos específicos.

*   **Fluxo de Operação:**
    1.  **Endpoint de API:** Uma API RESTful seria desenvolvida e implantada (e.g., usando Flask/FastAPI com Gunicorn/Uvicorn em um contêiner Docker/Kubernetes, ou AWS Lambda/Azure Functions/Google Cloud Functions).
    2.  **Requisição:** Um sistema cliente (e.g., o e-commerce da TechGlow) enviaria uma requisição para a API contendo as características do produto para o qual a previsão é solicitada.
    3.  **Geração de Features em Tempo Real:** A API receberia os dados de entrada, realizaria a engenharia de features necessária (com base em dados atualizados de sistemas transacionais ou cachê), e transformaria os dados (normalização) conforme o treinamento do modelo.
    4.  **Inferência do Modelo:** O modelo previamente carregado na memória da API realizaria a predição.
    5.  **Resposta:** A previsão seria retornada ao sistema cliente via API.

*   **Vantagens:** Previsões instantâneas, flexibilidade para integrar com aplicações interativas, permite personalização e tomada de decisão dinâmica.
*   **Desvantagens:** Maior complexidade de desenvolvimento e manutenção, custos operacionais potencialmente mais altos, necessidade de baixa latência e alta disponibilidade.

#### Escolha da Estratégia para TechGlow:

Para a **otimização de estoque e planejamento de compras**, a **Batch Prediction** é a estratégia principal e mais adequada, pois as decisões de estoque geralmente são tomadas em ciclos diários ou semanais. Ela atenderá os objetivos SMART do projeto de forma eficiente.

No entanto, para **usos futuros e mais avançados**, como sugestões de produtos personalizadas, precificação dinâmica ou alertas de fraude, a **Real-time Prediction via API** seria um complemento valioso. Inicialmente, o foco seria no batch, com a possibilidade de evoluir para tempo real conforme as necessidades de negócio e a maturidade da solução.

### 5. Considerações de Escalabilidade da Solução

A escalabilidade é um aspecto fundamental para garantir que a solução de previsão de demanda da TechGlow possa crescer e se adaptar às futuras necessidades do negócio, ao aumento do volume de dados e à complexidade dos modelos.

#### A. Escalabilidade para Volume de Dados:

*   **Armazenamento Distribuído e Elástico:** A utilização de Data Lakes (e.g., Amazon S3, Azure Data Lake Storage) e Data Warehouses baseados em nuvem (e.g., Google BigQuery, Amazon Redshift, Snowflake) garante escalabilidade ilimitada de armazenamento. Esses serviços são projetados para lidar com petabytes de dados, escalando automaticamente conforme a demanda.
*   **Processamento Distribuído:** Ferramentas de ETL/ELT como Apache Spark (executado em AWS EMR, Azure Databricks ou Google Dataproc) ou serviços gerenciados como AWS Glue, Azure Data Factory, são inerentemente escaláveis. Eles podem processar grandes volumes de dados em paralelo, distribuindo a carga de trabalho por múltiplos nós e expandindo a capacidade de computação conforme necessário.
*   **Bancos de Dados NoSQL:** Para o armazenamento de previsões que exigem alta performance de leitura e escrita (especialmente para cenários de tempo real), bancos de dados NoSQL (e.g., Amazon DynamoDB, Azure Cosmos DB, Google Cloud Firestore) podem ser utilizados, oferecendo escalabilidade horizontal e baixa latência.

#### B. Escalabilidade para Complexidade dos Modelos:

*   **Plataformas de Machine Learning em Nuvem:** Serviços como Amazon SageMaker, Google AI Platform e Azure Machine Learning oferecem infraestrutura gerenciada para treinamento e deployment de modelos. Isso inclui:
    *   **Recursos de Computação Flexíveis:** Possibilidade de provisionar VMs com diferentes configurações (CPUs, GPUs, memória) para treinar modelos complexos em tempo hábil.
    *   **Treinamento Distribuído:** Capacidade de treinar modelos em clusters distribuídos, acelerando o processo para conjuntos de dados maiores ou modelos mais complexos.
    *   **AutoML e Tunagem de Hiperparâmetros:** Ferramentas que automatizam a seleção de modelos e a otimização de hiperparâmetros, permitindo explorar modelos mais sofisticados sem exigir um esforço manual proporcionalmente maior.
*   **Versionamento e Gerenciamento de Modelos:** Plataformas de MLOps (parte das plataformas de ML em nuvem ou ferramentas como MLflow) permitem versionar modelos, rastrear experimentos e gerenciar o ciclo de vida do modelo, facilitando a atualização e melhoria contínua sem interrupções.

#### C. Escalabilidade para Demanda por Previsões Futuras:

*   **Serviços Serverless para Ingestão e Inferência:** Funções serverless (AWS Lambda, Azure Functions, Google Cloud Functions) são ideais para endpoints de API de previsão em tempo real. Elas escalam automaticamente de zero a milhares de requisições por segundo, pagando apenas pelo uso, sem a necessidade de gerenciar servidores.
*   **Orquestração Escalável:** Orquestradores como Apache Airflow (ou suas versões gerenciadas na nuvem como MWAA, ADF, GCP Composer) podem gerenciar milhares de tarefas e pipelines, garantindo que os processos de batch prediction sejam executados de forma confiável e escalável.
*   **Contêineres e Kubernetes:** Para deployment de modelos em tempo real que exigem mais controle ou funcionalidades específicas, a implantação em contêineres (Docker) e orquestração via Kubernetes (EKS, AKS, GKE) oferece alta disponibilidade, escalabilidade horizontal e gerenciamento eficiente de recursos.

Ao adotar uma arquitetura nativa da nuvem e utilizar serviços gerenciados e escaláveis, a TechGlow garante que sua solução de previsão de demanda possa crescer sem gargalos, suportando a expansão do negócio e as crescentes necessidades de análise e inteligência.

## Considerações de Ética e Governança de Dados

### Análise e Abordagem dos Aspectos Cruciais

Neste projeto de previsão de demanda para otimização de estoque, é fundamental abordar as considerações éticas e de governança de dados para garantir uma implementação responsável e sustentável da solução de IA.

### 1. Implicações Éticas e de Privacidade dos Dados

Os dados utilizados no projeto, especialmente `ID do Produto/SKU`, `Quantidade Vendida`, `Preço Unitário de Venda`, `Promoção/Desconto`, `Categoria do Produto` e, em um cenário real, `ID do Cliente` e `Local da Venda/Loja`, contêm informações que, se não tratadas adequadamente, podem levantar preocupações de privacidade. Embora nosso dataset sintético não contenha dados diretamente identificáveis de clientes, em um ambiente real, o histórico de compras de um `ID do Cliente` pode revelar padrões de consumo sensíveis ou permitir a reidentificação de indivíduos.

**Como a solução garante a privacidade dos usuários:**
*   **Anonimização/Pseudonimização:** Para dados de clientes (como `ID do Cliente`), a principal estratégia seria a pseudonimização. Isso significa substituir identificadores diretos por um código único gerado artificialmente. Em análises de massa, a anonimização (remoção irreversível de qualquer link para o indivíduo) seria aplicada a dados agregados. O `ID do Produto/SKU` e a `Categoria do Produto` não são considerados dados pessoais, mas seu uso combinado com outros dados pode gerar inferências.
*   **Princípio da Minimização de Dados:** Coletar e utilizar apenas os dados estritamente necessários para o objetivo da previsão de demanda. Por exemplo, informações demográficas muito detalhadas do cliente podem não ser essenciais para o modelo principal, mas podem ser consideradas para análises específicas e com consentimento.
*   **Controle de Acesso Rigoroso:** Implementar políticas de controle de acesso baseadas em função (Role-Based Access Control - RBAC) para garantir que apenas pessoal autorizado tenha acesso aos dados brutos e que esses acessos sejam logados e auditados.

### 2. Potencial de Viés Algorítmico

O viés algorítmico pode surgir se os dados históricos de vendas refletirem desigualdades sociais, padrões de consumo enviesados ou estratégias de marketing anteriores que favoreceram ou desfavoreceram determinados grupos ou produtos. Por exemplo:
*   **Viés de Disponibilidade:** Se historicamente certos produtos estavam frequentemente em falta ou não eram distribuídos equitativamente para todas as regiões/demografias de clientes, o modelo pode aprender que a demanda é baixa para esses produtos ou regiões, perpetuando o problema.
*   **Viés de Recomendação/Promoção:** Se promoções foram direcionadas predominantemente a um certo tipo de cliente ou para um tipo de produto, o modelo pode superestimar a demanda desses itens ou subestimar a demanda de outros, criando um ciclo de reforço.
*   **Viés de Gênero/Idade/Localização:** Embora nosso dataset não inclua essas variáveis, em dados reais, se a TechGlow tem um histórico de vendas concentrado em um determinado perfil demográfico, o modelo pode ter dificuldades em prever a demanda para outros segmentos de clientes, levando a decisões de estoque que não atendem a uma base de clientes mais ampla.

**Estratégias para mitigar ou monitorar vieses:**
*   **Análise de Viés nos Dados:** Realizar uma análise exploratória de dados focada em identificar desequilíbrios ou padrões injustos nos dados históricos antes do treinamento do modelo.
*   **Métricas de Fairness:** Utilizar métricas de fairness (ex: paridade demográfica, igualdade de oportunidades) durante a avaliação do modelo para garantir que ele performe de forma similar para diferentes segmentos de produtos ou grupos de clientes (se aplicável e identificado).
*   **Reamostragem/Ponderação:** Aplicar técnicas de reamostragem ou ponderação nos dados de treinamento para balancear representações sub-representadas ou corrigir distorções históricas.
*   **Interpretabilidade do Modelo (XAI):** Usar ferramentas de eXplainable AI (XAI) para entender quais features estão mais influenciando as previsões e se alguma delas está correlacionada com vieses indesejados.
*   **Monitoramento Contínuo:** Implementar um sistema de monitoramento pós-implantação para detectar qualquer viés emergente nas previsões e seu impacto nas operações de estoque.

### 3. Conformidade com Regulamentações de Proteção de Dados (LGPD)

A Lei Geral de Proteção de Dados (LGPD) no Brasil é altamente relevante para este projeto, especialmente se dados de clientes são utilizados. Os princípios-chave da LGPD incluem:
*   **Finalidade:** Os dados devem ser coletados para propósitos legítimos, específicos e informados ao titular. A previsão de demanda para otimização de estoque é uma finalidade legítima.
*   **Adequação e Necessidade:** A coleta e o processamento de dados devem ser compatíveis com a finalidade informada e limitados ao mínimo necessário. Isso reforça a importância da minimização de dados.
*   **Livre Acesso, Qualidade dos Dados:** Garantir que os titulares dos dados tenham acesso claro e facilitado aos seus dados e que estes sejam precisos e atualizados.
*   **Segurança e Prevenção:** Utilizar medidas técnicas e administrativas aptas a proteger os dados pessoais de acessos não autorizados e de situações acidentais ou ilícitas de destruição, perda, alteração, comunicação ou difusão.
*   **Não Discriminação:** O tratamento dos dados não pode ter fins discriminatórios ilícitos ou abusivos.
*   **Responsabilização e Prestação de Contas:** Demonstrar a adoção de medidas eficazes e capazes de comprovar a observância e o cumprimento das normas de proteção de dados pessoais.

**Como a solução se alinha à LGPD:**
*   O projeto adere à **Finalidade** de otimização de estoque. Qualquer uso de dados pessoais seria estritamente para essa finalidade ou outras claramente comunicadas.
*   A estratégia de **Minimização de Dados** e pseudonimização/anonimização demonstra **Adequação e Necessidade**.
*   A implementação de **Controles de Acesso** e medidas de segurança de dados (criptografia, firewalls, etc.) garante a **Segurança**.
*   A transparência sobre o uso dos dados e a possibilidade de os titulares exercerem seus direitos (acesso, correção, exclusão) são essenciais para a **Livre Acesso** e **Qualidade dos Dados**.
*   A análise e mitigação de viés algorítmico contribui para o princípio da **Não Discriminação**.

### 4. Plano de Governança de Dados

Um plano robusto de governança de dados é essencial para a sustentabilidade e a conformidade do projeto:

*   **Política de Coleta de Dados:**
    *   Definir quais dados são coletados, de onde, por que e por quanto tempo.
    *   Garantir o consentimento explícito dos titulares dos dados, quando necessário, e informar sobre o propósito da coleta.
    *   Estabelecer diretrizes para a qualidade e a consistência dos dados desde a origem.

*   **Política de Armazenamento de Dados:**
    *   Definir locais de armazenamento seguros (em conformidade com a LGPD, preferencialmente em território nacional para certos tipos de dados).
    *   Implementar criptografia para dados em repouso e em trânsito.
    *   Estabelecer políticas de retenção de dados, definindo por quanto tempo os dados serão armazenados e quando devem ser descartados.

*   **Política de Uso de Dados:**
    *   Especificar os propósitos legítimos para o uso dos dados (ex: treinamento de modelos, monitoramento, auditoria).
    *   Definir os níveis de acesso a diferentes tipos de dados para diferentes papéis dentro da organização (RBAC).
    *   Proibir o uso de dados para fins não autorizados ou discriminatórios.
    *   Exigir revisões regulares do uso de dados e conformidade.

*   **Política de Descarte de Dados:**
    *   Estabelecer procedimentos seguros para a eliminação de dados que não são mais necessários ou para os quais o período de retenção expirou, garantindo que não sejam recuperáveis.

*   **Responsabilidades e Papéis:**
    *   **Comitê de Governança de Dados:** Um grupo multifuncional (jurídico, TI, ciência de dados, negócio) responsável por definir e supervisionar as políticas de governança.
    *   **Data Protection Officer (DPO):** Responsável por garantir a conformidade com a LGPD e atuar como ponto de contato para autoridades e titulares dos dados.
    *   **Proprietários de Dados:** Responsáveis pela qualidade e integridade de conjuntos de dados específicos.
    *   **Usuários de Dados:** Responsáveis por aderir às políticas de uso e segurança dos dados.

*   **Auditoria e Monitoramento:**
    *   Realizar auditorias regulares das práticas de dados e conformidade com as políticas.
    *   Monitorar o desempenho do modelo em relação a vieses e fairness.
    *   Estabelecer um plano de resposta a incidentes de segurança de dados.

Este plano garante que o projeto não apenas atinja seus objetivos de negócio, mas o faça de forma ética, transparente e em conformidade com as regulamentações vigentes, construindo confiança e minimizando riscos.
